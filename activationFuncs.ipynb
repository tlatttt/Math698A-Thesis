```python
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np

def sigmoid(z):
    return 1 / (1 + np.exp(-z))

def relu(z):
    return np.maximum(0, z)

def derivative(f, z, eps=0.000001):
    return (f(z + eps) - f(z - eps))/(2 * eps)

def axes_settings(fig, ax, title, xlim, ylim):
    ax.set_xlim(xlim)
    ax.set_ylim(ylim)
    ax.set_title(title)
    
x = np.linspace(-5, 5, 200)

fig, axes = plt.subplots(1, 4, figsize=(16, 3))
axes_settings(fig, axes[0], "LTU function and its derivative", [-5, 5], [-1.2, 1.2])
axes[0].plot(x, np.sign(x), "k-", linewidth=1, label=r"$f(z)= sign(z)$")
axes[0].plot(x, derivative(np.sign, x), "b-", linewidth=1, label="Derivative")
axes[0].plot(0, 0, "rx", markersize=10)
axes[0].legend(loc=4)

axes_settings(fig, axes[1], "Logistic function and its derivative", [-5, 5], [-1.2, 1.2])
axes[1].plot(x, sigmoid(x), "k-", linewidth=2, label=r"$f(z) = \frac{1}{1 + e^{-z}}$")
axes[1].plot(x, derivative(sigmoid, x), "b-", linewidth=2, label="Derivative")
axes[1].legend(loc=3)

axes_settings(fig, axes[2], "Tanh function and its derivative", [-5, 5], [-1.2, 1.2])
axes[2].plot(x, np.tanh(x), "k-", linewidth=2, label=r"$f(z) = \frac{e^{2x}-1}{e^{2x}+1}$")
axes[2].plot(x, derivative(np.tanh, x), "b-", linewidth=2, label="Derivative")
axes[2].legend(loc=4)

axes_settings(fig, axes[3], "ReLU function and its derivative", [-5, 5], [-1.2, 1.2])
axes[3].plot(x, derivative(relu, x), "b-", linewidth=2, label="Derivative")
axes[3].plot(x, relu(x), "k-", linewidth=2, label=r"$f(z) = maz(0,z)$")
axes[3].plot(0, 0, "rx", markersize=10)
axes[3].legend(loc=4)
fig.savefig("activationFuncs.png")
```
